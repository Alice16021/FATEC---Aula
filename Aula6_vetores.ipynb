### 📌 Características das listas em Python
- São **mutáveis** (podem ser modificadas).
- Os elementos são **indexados**, a partir do índice `0`.
- Podem conter **tipos mistos** de dados.

```python
# Lista com diferentes tipos de dados
mistura = [10, "Olá", True, 3.14]
```

### 📥 Acessando elementos

```python
frutas = ["maçã", "banana", "uva", "laranja"] #começa com zero 0, então o índice (lugar) da maçã é 0.
print(frutas[0])  # maçã
print(frutas[2])  # uva
print(frutas[-1]) # laranja (último elemento)
```

### ✏️ Alterando elementos

```python
frutas[1] = "pera"
print(frutas)
```

### ➕ Adicionando elementos

```python
frutas.append("melancia")
print(frutas)

frutas.insert(2, "abacaxi")
print(frutas)
```

### ➖ Removendo elementos

```python
frutas.remove("uva")
print(frutas)

frutas.pop(0)
print(frutas)
```

### 🔄 Percorrendo listas com `for`

```python
numeros = [1, 2, 3, 4, 5]
for numero in numeros:
    print(numero * 2)
```

### 📐 Tamanho da lista

```python
numeros = [10, 20, 30]
print(len(numeros))
```

### 📚 Exercícios práticos

```python
alunos = ["Ana", "Bruno", "Carlos", "Diana", "Eduardo"]
for nome in alunos:
    print(nome, "está presente?")
```

```python
numeros = [5, 8, 12, 3, 7]
print("Primeiro:", numeros[0])
print("Último:", numeros[-1])
print("Soma:", sum(numeros))
```

```python
numeros = []
for i in range(5):
    valor = int(input(f"Digite o {i+1}º número: "))
    numeros.append(valor)

for n in numeros:
    print(n * 2)
```

```python
valores = [10, 25, 30, 40]
valores.remove(25)
valores.append(100)
print(valores)
```

### 🧠 Desafio

```python
produtos = []
while True:
    nome = input("Digite o nome do produto (ou 'sair' para encerrar): ")
    if nome.lower() == "sair":
        break
    produtos.append(nome)

print("Produtos cadastrados:")
for produto in produtos:
    print("-", produto)
```

### ✅ Conclusão
As listas são uma das estruturas mais poderosas e flexíveis do Python. Você usará listas frequentemente ao trabalhar com dados, coleções e estruturas mais complexas.

### 🧪 Exemplos adicionais de uso de listas

```python
# Criando uma lista de 0 a 9 automaticamente
lista = list(range(10))
print(lista)
```

```python
# Somando todos os elementos de uma lista
valores = [10, 20, 30]
total = sum(valores)
print("Total:", total)
```

```python
# Verificando se um elemento está na lista
nomes = ["Ana", "Beatriz", "Carlos"]
print("Carlos" in nomes)
print("João" in nomes)
```

```python
# Ordenando uma lista
notas = [8.5, 9.0, 6.5, 7.2]
notas.sort()
print(notas)
notas.reverse()
print(notas)
```

```python
# Copiando uma lista
original = [1, 2, 3]
copia = original[:]
copia.append(4)
print("Original:", original)
print("Cópia:", copia)
```
